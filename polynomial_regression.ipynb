```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(1)
x=2*np.random.rand(100,1)
```

```python
x
```

```python
#generating synthetic dp y
y = 4 +3*x+1.5*x**2+np.random.randn(100, 1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
#perform polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feature=PolynomialFeatures(degree=2,include_bias=True)
poly_feature
```

```python
x_poly_train=poly_feature.fit_transform(x_train)
x_poly_train
```

```python
x_poly_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_poly_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

Recall the polynomial model form (degree = 2)

Because you used PolynomialFeatures(degree=2), the model is:

y=βo+β1x+β2x^2

🔹 Substitute your learned values

β0=4.41704447 (from intercept_)

β1=1.86158348

β2=1.99655192

```python

x_test_poly=poly_feature.transform(x_test)
```

```python
y_pred=model.predict(x_test_poly)
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(y_test, y_pred)
print(mape * 100)
```
