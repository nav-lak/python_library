```python
import numpy as np
import matplotlib.pyplot as plt

x = [-360, -270, -180, -90, 0, 90, 180, 270, 360]

x_rad = [i * np.pi / 180 for i in x]

plt.plot(x, np.sin(x_rad), label="sin")
plt.plot(x, np.cos(x_rad), label="cos")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-360, 360, 1000)   # many points
x_rad = np.deg2rad(x)
plt.figure(figsize=(10, 5))   # ← increase figure size
plt.plot(x, np.sin(x_rad), label="sin")
plt.plot(x, np.cos(x_rad), label="cos")
plt.grid()

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-360, 360, 2000)
x_rad = np.deg2rad(x)

tan = np.tan(x_rad)
cot = 1 / tan

# Remove asymptotes
tan[np.abs(tan) > 15] = np.nan
cot[np.abs(cot) > 15] = np.nan

plt.figure(figsize=(10, 5))
plt.plot(x, tan, label="tan")
plt.plot(x, cot, label="cot")

plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-360, 360, 2000)
x=list(x)
x
x_rad=[i for i in x]
x_rad = [i * np.pi / 180 for i in x]

tan = np.tan(x_rad)
cot=1/tan
tanx=[]
cotx=[]
for i in list(tan):
  if -15 < i < 15:
    tanx.append(i)
  else :
    tanx.append(None)
for  j in list(cot):
  if -15<j<15:
    cotx.append(j)
  else:
    cotx.append(None)

plt.figure(figsize=(10, 5))
plt.grid()
plt.plot(x, tanx, label="tan")
plt.plot(x,cotx,label="cot")
```
